```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.cross_decomposition import PLSRegression
import plotly.express as px
import os
```

```python
def singleTissueClusterPlot(df,tissue,column,sheet):
    df_tissue = df[df['condition'].str.contains(tissue)]
    indexes = [sheet+'_'+tissue+'_15',sheet+'_'+tissue+'_30',sheet+'_'+tissue+'_60',sheet+'_'+tissue+'_120']
    toPlot = pd.DataFrame(index=indexes,
                          columns = list(set(df_tissue['species'])))
    for index,row in df_tissue.iterrows():
        condition = row['condition']
        species = row['species']
        value = row[column]
        toPlot.at[condition,species] = value
    toPlot = toPlot.fillna(0)
    toPlot = toPlot.loc[:, (toPlot != 0).any(axis=0)]
    return toPlot
```

### Generate heatmaps for all three estimates for HCC and Control 

```python
sheets = ['HCC','CTRL']
columns = ['sum_abs.3','Labelled_fraction.3','Labelled_absolute.3']
for sheet in sheets:
    df = pd.read_excel('data/210511_allM_editing_averages.xlsx',sheet_name=sheet)
    df = df[['condition', 'species', 'sum_abs.3','Labelled_fraction.3','Labelled_absolute.3']].copy()
    for column in columns:
        conditions = []
        for condition in set(df['condition']):
            condition = condition.split('_')[1]
            conditions.append(condition)
        conditions = list(set(conditions))
        dfs = []
        for condition in conditions:
            toPlot = singleTissueClusterPlot(df,condition,column,sheet)
            toPlot = toPlot.T
            dfs.append(toPlot)
            fig = sns.clustermap(toPlot,col_cluster=False,cmap='vlag',
                                 center=0,figsize=(5,8),yticklabels=True,xticklabels=True)
            fig.savefig('Figures/raw/'+column.replace('.3','')+'/'+sheet+'/'+condition+'.pdf')
        combined_df = pd.concat(dfs,axis=1)
        combined_df = combined_df.fillna(0)
        fig = sns.clustermap(combined_df,col_cluster=False,cmap='vlag',center=0,yticklabels=True,xticklabels=True)
        fig.savefig('Figures/raw/'+column.replace('.3','')+'/'+sheet+'/combined.pdf')
```

```python
CTRL = pd.read_excel('data/210511_allM_editing_averages.xlsx',sheet_name='CTRL')
CTRL = CTRL[['condition', 'species', 'sum_abs.3','Labelled_fraction.3','Labelled_absolute.3']].copy()
HCC = pd.read_excel('data/210511_allM_editing_averages.xlsx',sheet_name='HCC')
HCC = HCC[['condition', 'species', 'sum_abs.3','Labelled_fraction.3','Labelled_absolute.3']].copy()
```

```python
df = pd.DataFrame()
for index,row in HCC.iterrows():
    df.at[row['species'],row['condition']] = row['sum_abs.3']
for index,row in CTRL.iterrows():
    df.at[row['species'],row['condition']] = row['sum_abs.3']
df = df.fillna(0)
df = df.T
```

```python
sample_annot = pd.DataFrame(columns=['Group','Group_number'])
sample_annot['Group'] = list(set(df.index))
sample_annot['Group_number'] = list(np.arange(1,len(set(df.index))+1))
```

```python
ncomps = 10
plsr = PLSRegression(n_components=ncomps, scale=False)
plsr.fit_transform(df,sample_annot['Group_number'])
principalComponents = plsr.x_scores_
variance_in_x = np.var(plsr.x_scores_, axis = 0) 
total_variance_in_x = np.var(df.to_numpy(), axis = 0)
fractions_of_explained_variance = variance_in_x / sum(total_variance_in_x)
fractions_of_explained_variance = fractions_of_explained_variance * 100
fractions_of_explained_variance = np.round_(fractions_of_explained_variance,decimals=2)

columns = []
count = 1
while count <= ncomps:
    columns.append('Component '+str(count)+' ('+str(fractions_of_explained_variance[count-1])+'%)')
    count += 1
    
principalDf = pd.DataFrame(data = principalComponents,columns=columns)
finalDf = pd.concat([principalDf, sample_annot[['Group']]], axis = 1)
```

```python
finalDf['Experiment'] = finalDf['Group'].str.split('_').str[0]
finalDf['Tissue'] = finalDf['Group'].str.split('_').str[1]
finalDf['Time'] = finalDf['Group'].str.split('_').str[2]
```

```python
fig = px.scatter(finalDf, 
                 x = finalDf.columns[0], 
                 y = finalDf.columns[1],
                 color = 'Tissue',
                 template = 'none',)
fig = fig.update_layout(autosize = False,
                        height = 1000,
                        width = 1000,
                        font_family="Arial",
                        font_color="black",
                        font_size=12)
fig.update_traces(mode='markers', 
                  marker_line_width=2, 
                  marker_size=14)
#fig.write_image('Figures/PLSDA_withPBQC.pdf')
fig.show()
```

```python
df = pd.concat([HCC,CTRL],axis=0)
df['Experiment'] = df['condition'].str.split('_').str[0]
df['Tissue'] = df['condition'].str.split('_').str[1]
df['Time'] = df['condition'].str.split('_').str[2]
df.head()
```

```python
plotTypes = ['Labelled_absolute.3','sum_abs.3','Labelled_fraction.3']
for plotType in plotTypes:
    print(plotType)
    newdir = 'Figures/lineplots/'+plotType.replace('.3','')+'/'
    os.system('mkdir '+newdir)
    toPlot = df.dropna()
    for species in set(df['species']):
        toPlot = df[df['species'] == species]
        fig = px.line(toPlot,
                      x='Time',
                      y=plotType,
                      facet_col='Experiment',
                      facet_row='Tissue',
                      color='species',
                      markers=True,
                      template='none')
        fig.update_layout(autosize=False,height=1200,width=900)
        fig.update_xaxes(tick0=0,dtick=15)
        fig.write_image(newdir+species+'.pdf')
```
